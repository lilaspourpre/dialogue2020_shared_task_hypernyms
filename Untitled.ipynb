```python
import json
import random
import numpy as np
```

```python
positive = []
negative = []
with open("data_vectors_large_new_proper.jsonlines", 'r', encoding='utf-8') as f:
    for line in f:
        data_line = json.loads(line)
        if data_line['label'] == 1:
            positive.append(data_line)
        else:
            negative.append(data_line)
len(positive), len(negative)
```

```python
negative = random.sample(negative, len(positive))
new_dataset = positive + negative
```

```python
with open("data_vectors_balanced_new.jsonlines", 'w', encoding='utf-8', newline='\n') as w:
    for dataset in new_dataset:
        w.write(json.dumps(dataset)+"\n")
```

```python
vectors = np.array([np.array(line['vector']) for line in new_dataset])
vectors.shape
```

```python
labels = np.array([line['label'] for line in new_dataset])
labels.shape
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
clf = LogisticRegression(random_state=0).fit(vectors, labels)
```

```python
clf.score(vectors, labels)
```
