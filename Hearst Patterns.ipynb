```python
from pymystem3 import Mystem
import gzip
from collections import defaultdict
```

```python
m = Mystem()
def lemmatize_text(text):
    lemmas = m.lemmatize(text)[:-1]
    lemmatized_text = ''.join(lemmas)
    return lemmatized_text
```

```python
nouns_path = "baselines/ruwordnet_non-restricted-nouns_no_labels_final.tsv"
verbs_path = "baselines/ruwordnet_non-restricted-verbs_no_labels_final.tsv"
```

```python
with open(nouns_path, 'r', encoding="utf-8") as f:
    nouns = f.read().lower().split("\n")
    
with open(verbs_path, 'r', encoding="utf-8") as f:
    verbs = f.read().lower().split("\n")
```

```python
words = defaultdict(list)
with gzip.open("ru-librusec-wiki-diff.csv.gz", 'rt', encoding="utf-8") as f:
    f.readline()
    for line in f:
        word, hypernym, _ = line.strip().lower().split("\t")
        for w in lemmatize_text(word).split():
            if w in nouns or w in verbs:
                words[w].append(lemmatize_text(hypernym))
```

```python
f.close()
```
